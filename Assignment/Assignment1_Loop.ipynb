```python
# ==========================================================
# Question 1: ATM PIN Verification
#
# Problem:
# A user gets a maximum of 3 attempts to enter the correct
# ATM PIN. If the entered PIN matches the correct PIN,
# display "Access Granted" and stop checking further attempts.
# Otherwise, display the number of remaining attempts.
# If all attempts fail, display "Card Locked".
#
# Approach:
# • Used enumerate() to get both the attempt number and PIN.
# • Used list slicing ([:3]) to allow only the first 3 attempts.
# • Calculated remaining attempts after every incorrect PIN.
# • Used if-elif-else for decision making.
# • Used break to stop the loop immediately after a successful login.
# ==========================================================
# USING FOR LOOP
# --- FOR LOOP ---
# Question 1: ATM PIN Verification
correct_pin = "4521"
entered_pins = ["1111","2222","4521"]

for i, pin in enumerate(entered_pins[:3]):
    tries_left=3-(i+1)
    if pin == correct_pin:
        print("Access Granted")
        break
    elif tries_left > 0:
        print(f"Wrong PIN, {tries_left} tries left")
    else:
        print ("Card Blocked")



```

```python
#Question 1: ATM PIN Verification
# USING WHILE LOOP
#
# Approach:
# - Use an index variable to access each PIN in the list.
# - Continue while the index is within the list and fewer
#   than three attempts have been processed.
# - Calculate the number of remaining attempts.
# - Stop the loop when the correct PIN is found.
# ==========================================================

correct_pin = "4521"
entered_pins = ["1111", "2222", "4521"]

max_attempts = 3
i = 0

while i < len(entered_pins) and i < max_attempts:
    pin = entered_pins[i]
    tries_left = max_attempts - (i + 1)

    if pin == correct_pin:
        print("Access Granted")
        break

    elif tries_left > 0:
        print(f"Wrong PIN, {tries_left} tries left")

    else:
        print("Card Locked")

    i += 1
```

```python

# Question 1: ATM PIN Verification
# IF pins are manually entered
# Approach:
# - Ask the user to manually enter the PIN.
# - Continue until the correct PIN is entered or all
#   three attempts are used.
# - Display the remaining attempts after each wrong PIN.

correct_pin = "4521"
attempts = 3

while attempts > 0:
    entered_pin = input("Enter your PIN: ")

    if entered_pin == correct_pin:
        print("Access Granted")
        break

    else:
        attempts -= 1

        if attempts > 0:
            print("Wrong PIN,", attempts, "tries left")
        else:
            print("Card Locked")
```

```python
# ==========================================================
# Question 2: Grocery Checkout Total — FOR Loop
#
# Problem:
# Add grocery item prices to a running total.
# Once the total exceeds ₹1000, apply a 10% discount
# to all remaining items.
#
# Display:
# - Original Total
# - Total Discount
# - Final Amount
# ==========================================================

item_prices = [300, 250, 500, 200, 100]

#item_prices = [50,100,200,300,20]

original_total = 0
final_total = 0
total_discount = 0
discount_active = False

for price in item_prices:

    original_total += price

    if discount_active:
        discount = price * 0.10
        total_discount += discount
        final_total += price - discount

    else:
        final_total += price

        if final_total > 1000:
            discount_active = True

print("\n========== Grocery Bill ==========")
print(f"Original Total : ${original_total:.2f}")
print(f"Total Discount : ${total_discount:.2f}")
print("----------------------------------")
print(f"Final Amount   : ${final_total:.2f}")
print("==================================")

#print("Original Total: $", original_total)
#print("Total Discount: $", total_discount)
#print("Final Amount: $", final_total)
```

```python
# ==========================================================
# Question 3: Number Guessing Game — FOR Loop
#
# Problem:
# Compare guesses with a secret number.
# Print "Too Low", "Too High", or "Correct".
# Stop when the correct number is found and display
# the number of attempts.
#
# Approach:
# - Loop through every guess.
# - Increase the attempt counter for each guess.
# - Compare each guess with the secret number.
# - Use break when the correct answer is found.
# ==========================================================

secret_number = 8
guesses = [4, 10, 7, 8, 9]

attempts = 0

for guess in guesses:
    attempts += 1

    if guess < secret_number:
        print("Too Low")

    elif guess > secret_number:
        print("Too High")

    else:
        print("Correct")
        print(f"Guessed in {attempts} attempts")
        break
```

```python
# ==========================================================
# Question 3: Number Guessing Game — WHILE Loop
#
# Problem:
# Compare guesses with a secret number and stop when
# the correct number is found.
#
# Approach:
# - Use an index to access guesses one at a time.
# - Compare each guess using if-elif-else.
# - Use i + 1 as the attempt number because indexing
#   starts at zero.
# - Stop the loop using break when the guess is correct.
# ==========================================================

secret_number = 8
guesses = [4, 10, 7, 8, 9]

i = 0

while i < len(guesses):
    guess = guesses[i]

    if guess < secret_number:
        print("Too Low")

    elif guess > secret_number:
        print("Too High")

    else:
        print("Correct")
        print(f"Guessed in {i + 1} attempts")
        break

    i += 1
```

```python
# ==========================================================
# Question 3: Number Guessing Game — WHILE Loop
#
# Input numbers manually
# The user gets a maximum of five attempts.
# Approach:
# - Ask the user to enter a guess.
# - Compare the guess with the secret number.
# - Display whether the guess is too high or too low.
# - Count the number of attempts.
# - Stop the loop when the correct number is guessed.
# ==========================================================

secret_number = 8
attempts = 0
max_attempts = 5

while attempts < max_attempts:

    guess = int(input("Enter your guess: "))
    attempts += 1

    if guess < secret_number:
        print("Too Low")

    elif guess > secret_number:
        print("Too High")

    else:
        print("Correct!")
        print(f"You guessed the number in {attempts} attempts.")
        break

else:
    print("Game Over!")
    print(f"The secret number was {secret_number}.")
```

```python
# ==========================================================
# Question 3: Number Guessing Game — FOR Loop
#
# Input numbers manually
# The user gets a maximum of five attempts.
#
# Approach:
# - Use a FOR loop to limit the number of attempts.
# - Ask for a guess during each iteration.
# - Compare the guess with the secret number.
# - Stop the loop when the correct number is guessed.
# ==========================================================

secret_number = 8
max_attempts = 5

for attempt in range(1, max_attempts + 1):

    guess = int(input("Enter your guess: "))

    if guess < secret_number:
        print("Too Low")

    elif guess > secret_number:
        print("Too High")

    else:
        print("Correct!")
        print(f"You guessed the number in {attempt} attempts.")
        break

else:
    print("Game Over!")
    print(f"The secret number was {secret_number}.")
```

```python
# ==========================================================
# Question 4: Movie Ticket Booking — FOR Loop
#
# Problem:
# Process booking requests until all seats are sold or
# a request asks for more seats than are available.
#
# Approach:
# - Store the available seat count.
# - Loop through every booking request.
# - Deduct the request when enough seats are available.
# - Stop if a request exceeds the available seats.
# - Display the remaining seats at the end.
# ==========================================================

total_seats = 20
booking_requests = [5, 3, 4, 8]

available_seats = total_seats

for request in booking_requests:

    if request <= available_seats:
        available_seats -= request
        print(f"Booked {request} seats")

        if available_seats == 0:
            print("All seats sold")
            break

    else:
        print("Not enough seats")
        break

print(f"Seats remaining: {available_seats}")
```

```python
# ==========================================================
# Question 4: Movie Ticket Booking — WHILE Loop
#
# Problem:
# Process booking requests one by one until the seats
# are sold out or a request cannot be fulfilled.
#
# Approach:
# - Use an index to access each booking request.
# - Check the request against the available seats.
# - Deduct successful bookings from the remaining seats.
# - Stop processing when there are not enough seats.
# ==========================================================

total_seats = 20
booking_requests = [5, 3, 4, 8]

available_seats = total_seats
i = 0

while i < len(booking_requests):
    request = booking_requests[i]

    if request <= available_seats:
        available_seats -= request
        print(f"Booked {request} seats")

        if available_seats == 0:
            print("All seats sold")
            break

    else:
        print("Not enough seats")
        break

    i += 1

print(f"Seats remaining: {available_seats}")
```

```python
# ==========================================================
# Question 5: Study Streak Tracker — FOR Loop
#
# Problem:
# Count consecutive quiz scores that are 60 or higher.
# Stop processing when a score falls below 60.
#
# Approach:
# - Start the streak counter at zero.
# - Loop through each score.
# - Increase the streak when the score is at least 60.
# - Stop the loop when the score is below 60.
# - Display the final streak length.
# ==========================================================

#scores = [80, 75, 90, 64, 55, 92]
scores = [65,75,90,85,60,50]

streak = 0

for score in scores:

    if score >= 60:
        streak += 1
        print(f"Score {score}: Day Passed")

    else:
        print(f"Score {score}: Streak Broken")
        break

print(f"Final streak length: {streak}")
```

```python
# ==========================================================
# Question 5: Study Streak Tracker — WHILE Loop
#
# Problem:
# Count consecutive quiz scores of 60 or above.
# Stop when a score below 60 is found.
#
# Approach:
# - Use an index to move through the score list.
# - Increase the streak for passing scores.
# - Stop the loop using break when the streak is broken.
# - Increase the index after each passing score.
# ==========================================================

#scores = [80, 75, 90, 64, 55, 92]
scores = [65,75,90,85,60,50]
streak = 0
i = 0

while i < len(scores):
    score = scores[i]

    if score >= 60:
        streak += 1
        print(f"Score {score}: Day Passed")

    else:
        print(f"Score {score}: Streak Broken")
        break

    i += 1

print(f"Final streak length: {streak}")
```
